```python
libraries_imported = 'alarmas_bsc_celdas_indispo.py,alarmas_bsc_celdas_captura.py'

specs = [imp.spec_from_file_location(library[:-3],library) for library in libraries_imported.split(',')]
modules =[(imp.module_from_spec(spec),spec) for spec in specs]
```

```python
for module,spec in modules:
    data = []
    spec.loader.exec_module(module)
    data.append(module.funcion())
```
